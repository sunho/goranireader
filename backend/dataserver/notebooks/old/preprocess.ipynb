```python
import pandas
import os
import json
import numpy as np
df = pandas.read_json('./data/logs.json')
```

```python
words = df['sentences'].map(lambda x: list(map(lambda y: y['words'], x))).explode().explode().reset_index()
words = words[['sentences']].rename(columns={'sentences': 'word'})
words = words['word'].map(lambda x: str(x).lower()).unique()
```

```python
df['wordLen'] = df['sentences'].map(lambda x: len([word for sentecne in x for word in sentecne['words']]))
```

```python
df2 = df.copy()
df2['i'] = np.arange(len(df2))
df3 = df2.copy()
df2 = df2.explode('sentences')
df2['signalWord'] = df2['sentences'].map(lambda x: [[x['words'][i].lower(), 0 if (i in x['unknownIndice']) or (x['words'][i].lower() in [y.lower() for y in x['unknownWords']]) else 1] for i in range(len(x['words']))])
df3['signalWord'] = df2.groupby('i')['signalWord'].apply(sum)
del df3['i']
df2 = df3
```

```python
from gogorani.timeutil import parse_ts
df2['time'] = df2['time'].map(lambda x: parse_ts(x))
```

```python
df2['wpm'] = df2['wordLen'] / (df2['eltime']/1000) * 60
df2 = df2[df2['wpm'] < 400]
```

```python
df5 = df2.loc[df2['userId'] == '8ACCQYz8NMc4Lc9Nze3R'].copy()
df5['sentences'] = df5['sentences'].map(lambda x: x[0]['sid'])
df5['signalWord'] = df5['signalWord'].map(lambda x: [y[0] for y in x if y[1] == 0])
df5.sort_values(['time'])[:50]
```

```python
df3 = df2.loc[df2['wpm'] < 400]
df3.groupby('userId').mean()
```

```python
def handle(value):
    words = []
    knows = []
    times = []
    for x, t in zip(value['signalWord'], value['time']):
        for y in x:
            if len(y) != 2:
                continue    
            words.append(y[0])
            knows.append(y[1])
            times.append(t)
    return pandas.DataFrame({'word': words, 'known': knows, 'time': times})
df3 = df2.groupby('userId').apply(handle).reset_index(level=1, drop=True).reset_index()
df3.set_index(['userId', 'word'], inplace=True)
df3 = df3.sort_values(['time'])
```

```python
def retention(df):
    count = 0
    start = 0
    total = 0
    days = []
    maxDay = 0
    
    for t, k in zip(df['time'], df['known']):
        if start == 0:
            start = t
        dur = t - start
        total += k
        count += 1
        tmp = int(dur/60)
        days.append({'day': tmp, 'known': k})
        if maxDay < tmp:
            maxDay = tmp
    df = pandas.DataFrame(days).groupby('day').mean().query('known != 0').reset_index()
    if df.size >=2:
        first = df.iloc[0]['known']
        last = df.iloc[-1]['known']
    else:
        first = np.nan
        last = np.nan
    corr = df.corr(method='pearson')
    corr = corr['day'].loc['known']
    return pandas.Series({'count': count, 'days': maxDay + 1, 'avg': total/count, 'corr': corr, 'first': first, 'last': last})

df4 = df3.loc['8ACCQYz8NMc4Lc9Nze3R'].groupby('word').apply(retention)

```
