```python
import pandas as pd
signal_df = pd.read_hdf('signal.h5')
page_df = pd.read_hdf('pages.h5')

del page_df['signal']
```

```python
signal_df = signal_df.groupby(['userId', 'session'])\
    .agg({'signal': 'mean'})\
    .reset_index()
signal_df
```

```python
page_df = page_df.set_index(['userId', 'session'])
signal_df = signal_df.set_index(['userId', 'session'])

df = page_df.join(signal_df, how='inner')\
    .reset_index()
```

```python
import plotly.express as px
fig = px.density_contour(df, x="wpm", y="signal")
fig.update_traces(contours_coloring="fill", contours_showlabels = True)
fig
```
